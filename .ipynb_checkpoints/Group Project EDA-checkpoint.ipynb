```python
# User Defined Function to drop columns as per the percentage provided

def deleteNulls(dataframe, axis =1, percent=0.3):
    '''
    * deleteNull function will remove the rows and columns based on parameters provided.
    * dataframe : Name of the dataframe  
    * axis      : axis = 0 defines drop rows, axis =1(default) defines drop columns    
    * percent   : percent of data where column/rows values are null,default is 0.3(30%)
              
    '''
    df = dataframe.copy()
    ishape = df.shape
    if axis == 0:
        rownames = df.transpose().isnull().sum()
        rownames = list(rownames[rownames.values > percent*len(df)].index)
        df.drop(df.index[rownames],inplace=True) 
        print("\nNumber of Rows dropped\t: ",len(rownames))
    else:
        colnames = (df.isnull().sum()/len(df))
        colnames = list(colnames[colnames.values>=percent].index)
        df.drop(labels = colnames,axis =1,inplace=True)        
        print("Number of Columns dropped\t: ",len(colnames))
        
    print("\nOld dataset rows,columns",ishape,"\nNew dataset rows,columns",df.shape)

    return df
```

```python
# Dropping columns where NULL values are equal to or greater  than 50%

loan = deleteNulls(loan, axis=1, percent = 0.5)
```

### Univariate Analysis

Boxplots, Histograms, Distribution Plot, Violin Plot etc are used for Visualizations

```python
# User defined Function to plot the variables for Univariate Analysis

def univariate(df,col,vartype,hue =None):
    
    '''
    Univariate function will plot the graphs based on the parameters.
    df      : dataframe name
    col     : Column name
    vartype : variable type : continuos or categorical
                Continuos(0)   : Distribution, Violin & Boxplot will be plotted.
                Categorical(1) : Countplot will be plotted.
    hue     : It's only applicable for categorical analysis.
    
    '''
    sns.set(style="darkgrid")
    
    
    if vartype == 0:
        fig, ax=plt.subplots(nrows =1,ncols=3,figsize=(20,8))
        ax[0].set_title("Distribution Plot")
        sns.distplot(df[col],ax=ax[0], color='r')
        ax[1].set_title("Violin Plot")
        sns.violinplot(data =df, x=col,ax=ax[1], inner="quartile", color='g')
        ax[2].set_title("Box Plot")
        sns.boxplot(data =df, x=col,ax=ax[2],orient='v',color='y')
    
    if vartype == 1:
        temp = pd.Series(data = hue)
        fig, ax = plt.subplots()
        width = len(df[col].unique()) + 6 + 4*len(temp.unique())
        fig.set_size_inches(width , 7)
        ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue) 
        if len(temp.unique()) > 0:
            for p in ax.patches:
                ax.annotate('{:1.1f}%'.format((p.get_height()*100)/float(len(loans))), (p.get_x()+0.12, p.get_height()+20))
        else:
            for p in ax.patches:
                ax.annotate(p.get_height(), (p.get_x()+0.32, p.get_height()+20)) 
        del temp
    else:
        exit
        
    plt.show()
```

## Charge Off Probability

As per the dataset, we will analyse the different factors and variables that led to charge off the loan. As the charge off leads to loss for the company.  Hence once the factors are determined, then the company can take proactive steps in analysing these factors in future before they grant loans. Also based on these factors, the company can decide to give loan or reject the loan.  The probability with respect to the charge off against different variables are given below.

Two User Defined Functions are defined here 
1. Crosstab fuction for Summarized Data
2. Probability function for columns to be analyzed

```python
# Crosstab User Defined Function
def crosstab(df,col):
    '''
    df : Dataframe
    col: Column Name
    '''
    crosstab = pd.crosstab(df[col], df['loan_status'],margins=True)
    crosstab['Charged Off Probability'] = round((crosstab['Charged Off']/crosstab['All']),3)
    crosstab = crosstab[0:-1]
    return crosstab
```

```python
# Probability User Defined Function for Charge Off

def prob_chargeoff(df,col,stacked= True):
    '''
    df      : Dataframe
    col     : Column Name
    stacked : True(default) for Stacked Bar
    '''
    # get dataframe from crosstab function
    plotCrosstab = crosstab(df,col)
    
    linePlot = plotCrosstab[['Charged Off Probability']]      
    barPlot =  plotCrosstab.iloc[:,0:2]
    ax = linePlot.plot(figsize=(20,8), marker='o',color = 'b')
    ax2 = barPlot.plot(kind='bar',ax = ax,rot=1,secondary_y=True,stacked=stacked)
    ax.set_title(df[col].name.title()+' vs Charged Off Probability',fontsize=20,weight="bold")
    ax.set_xlabel(df[col].name.title(),fontsize=14)
    ax.set_ylabel('Charged Off Probability',color = 'b',fontsize=14)
    ax2.set_ylabel('Number of Applicants',color = 'g',fontsize=14)
    plt.show()
```

```python
# Location and Charge off Probability
filter_states = loans.addr_state.value_counts()
filter_states = filter_states[(filter_states < 10)]

loan_filter_states = loans.drop(labels = loans[loans.addr_state.isin(filter_states.index)].index)
```

```python
states = crosstab(loan_filter_states,'addr_state')
display(states.tail(20))

prob_chargeoff(df =loan_filter_states,col ='addr_state')
```

#### Observation
It is observed that NV, NM , SD and NJ areas have high rates of Charged Off Loans.  Hence it is prudent to discover more why default is happening in these areas and company must take care of this in future disbursals.

```python
# Loan Purpose and Charge off Probability
loan_purpose = crosstab(loans,'purpose')
display(loan_purpose)

prob_chargeoff(df =loans,col ='purpose',stacked=False)
```

#### Observations
The small business loans, renewable energy loans and educational loans had the higher risk of loan defaults

```python
# Grade/Subgrade and Charge Off Probability
grade = crosstab(loans,'grade')
display(grade)

prob_chargeoff(df =loans,col ='grade',stacked=False)
prob_chargeoff(df =loans,col ='sub_grade')
```
